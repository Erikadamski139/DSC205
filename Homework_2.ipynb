```python
#DSC 101 Homework 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from os import stat
from scipy.stats import ttest_1samp
```

```python
# One sample t Test
popmean = 30
n = 25
sm = 27.5
std = 5
cl = 0.95
alpha = 0.05

ho = "Daily average is = 30"
ha = "Daily average is not = 30"

ttest = (sm - popmean)/(std/np.sqrt(n))
print("T-test", ttest)

df = n - 1

p_value = stats.t.sf(np.abs(ttest), df) * 2
print("P-value", p_value)

if p_value < alpha:
    print("Reject Null")
else:
    print("Do not reject Null")

t_c = stats.t.ppf(1 - alpha/2, df)
moe = t_c * (std/np.sqrt(n))
lb = sm - moe
ub = sm + moe
print("At the 95% condifence level:", lb, ub)
```

```python
#Paired T-test
before = np.array([145, 150, 155, 140, 148, 152, 147, 149, 151, 153])
After = np.array([138, 142, 147, 135, 140, 144, 139, 141, 143, 145])
n1 = 10
alpha = 0.05
df =  before - After

null = "Blood pressure After is = Blood Preasure before"
alt = "Blood Preasure after is > Blood preasure before"

t, p = stats.ttest_rel(before, After)
print("T-Test:", t)
print("P-Value:", p)

if p < alpha:
    print("Reject Null")
else:
    print("Do not reject Null")

mdf = np.mean(df)
std = np.std(df)
t_c = stats.t.ppf(1 - alpha/2, df = n1 - 1)
moe = t_c * (std/ np.sqrt(n1))
lb = mdf - moe
ub = mdf + moe
print("At the 95% condifence level:", lb, ub)
```

```python
#Two Sample T-Test
pa = np.array([3.2, 2.8, 3.5, 4.1, 3.9, 3.0, 4.5, 3.3, 3.8, 2.9])
pb = np.array([2.1, 2.5, 2.8, 2.4, 2.6, 2.3, 2.7, 2.9, 2.5, 2.8])

null = "There is no signifiant difference between the two programs"
alt = "Their is a signifigant difference"

alpha = 0.05

t, p = stats.ttest_ind(pa, pb)
print("T-test", t)
print("P-Value", p)

if p < alpha:
    print("Reject Null")
else:
    print("Do not reject the Null")

n1 = 10
n2 = 10
df = n1 + n2 - 2
std = np.sqrt(((n1 - 1 )* np.std(pa, ddof = 1)**2 + (n2 - 1)* np.std(pb, ddof = 1)**2) / (n1 + n2 - 2))
t_c = stats.t.ppf(1 - alpha/ 2, df = df)
moe = t_c * std * np.sqrt(1/n1 + 1/n2)
m = np.mean(pa) - np.mean(pb)
lb = m - moe
ub = m + moe
print("At the 95% condifence level:", lb, ub)
```
